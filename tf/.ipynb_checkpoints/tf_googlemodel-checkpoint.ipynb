## 加载inception模型

```python
inception_graph_def_file = os.path.join('./inception_model', 'classify_image_graph_def.pb')

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)

with tf.gfile.FastGFile(inception_graph_def_file, 'rb') as f:
    graph_def = tf.GraphDef()
    graph_def.ParseFromString(f.read())
    tf.import_graph_def(graph_def, name='inception')
    
    writer = tf.summary.FileWriter('./inception_log', sess.graph)
    writer.close()
sess.close()
```

## 使用inception模型

```python
class NodeLookup(object):
    def __init__(self):
        label_lookup_path = './inception_model/imagenet_2012_challenge_label_map_proto.pbtxt'
        uid_lookup_path = './inception_model/imagenet_synset_to_human_label_map.txt'
        self.node_lookup = self.load(label_lookup_path, uid_lookup_path)
        
    def load(self, label_lookup_path, uid_lookup_path):
        proto_as_ascii_lines = tf.gfile.GFile(uid_lookup_path).readlines()
        uid_to_human = {}
        for line in proto_as_ascii_lines:
            line = line.strip('\n')
            parsed_items = line.split('\t')
            uid = parsed_items[0]
            human_string = parsed_items[1]
            uid_to_human[uid] = human_string
            
        proto_as_ascii_lines = tf.gfile.GFile(label_lookup_path).readlines()
        node_id_to_uid = {}
        for line in proto_as_ascii_lines:
            if 'target_class:' in line:
                target_class = line.split(': ')[1]
            if 'target_class_string:' in line:
                target_class_string = line.split(': ')[1]
                node_id_to_uid[target_class[:-1]] = target_class_string[1:-2]
            
        node_id_to_name = {}
        for k, v in node_id_to_uid.items():
            node_id_to_name[k] = uid_to_human[v]
        
        return node_id_to_name
    
    def id_to_string(self, node_id):
        node_id = str(node_id)
        if node_id not in self.node_lookup:
            return ''
        else:
            return self.node_lookup[node_id]
```

```python
with tf.gfile.FastGFile('./inception_model/classify_image_graph_def.pb', 'rb') as f:
    graph_def = tf.GraphDef()
    graph_def.ParseFromString(f.read())
    tf.import_graph_def(graph_def, name='inception')
```

```python
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)

softmax_tensor = sess.graph.get_tensor_by_name('inception/softmax:0')

for root, dirs, files in os.walk('./inception_model/images/'):
    for file in files:
        image_data = tf.gfile.FastGFile(os.path.join(root, file), 'rb').read()
        prediction = sess.run(softmax_tensor, feed_dict={'inception/DecodeJpeg/contents:0':image_data})
        prediction = np.squeeze(prediction) # 转化为一维数据
        
        image_path = os.path.join(root, file)
        print(image_path)
        
        # 显示图片
        img = Image.open(image_path)
        plt.imshow(img)
        plt.axis('off')
        plt.show()
        
        # 排序
        top_k = prediction.argsort()[-5:][::-1]
        node_lookup = NodeLookup()
        for node_id in top_k:
            human_string = node_lookup.id_to_string(node_id)
            score = prediction[node_id]
            print(human_string + '(score = ' + str(score) + ')')
```
