```python
tr_split = pd.read_csv('tr_split_feature.csv')
tr_feature = pd.read_csv('tr_feature_feature.csv')
tr_single = pd.read_csv('tr_single_feature.csv')

tr_split.shape
tr_split.head()
tr_feature.shape
tr_feature.head()
tr_single.shape
tr_single.head()
```

```python
split_col = ['marriageStatus', 'interest1', 'interest2', 'interest3', 'interest4', 'interest5', 'kw1', 'kw2', 'kw3',
            'topic1', 'topic2', 'topic3', 'appIdInstall', 'appIdAction', 'ct', 'os']
feature_col = ['age', 'gender', 'marriageStatus', 'education', 'consumptionAbility', 'LBS', 'interest1', 'interest2',
               'interest3', 'interest4', 'interest5', 'kw1', 'kw2', 'kw3', 'topic1', 'topic2', 'topic3', 'appIdInstall',
               'appIdAction', 'ct', 'os', 'carrier','house']
single_col = ['age', 'gender', 'education', 'consumptionAbility', 'LBS', 'carrier','house']
```

```python
tr_single = tr_single.drop(single_col + ['aid', 'uid'], axis=1)
tr_split = tr_split.drop(feature_col + ['aid', 'uid', 'is_his', 'label'], axis=1)

tr_split.shape
tr_split.head()
tr_feature.shape
tr_feature.head()
tr_single.shape
tr_single.head()
```

```python
train = pd.concat([tr_feature, tr_split, tr_single], axis=1)
predictors = list(set(train.columns) - set(split_col))
train = train[predictors]

train.shape
train.head()

del tr_feature
del tr_split
del tr_single
```

```python
train = pd.read_csv('tr_processing55.csv')

train.shape
train.head()
```

```python
predictors = list(set(train.columns) - set(split_col) - set(['is_his', 'aid', 'uid', 'label']))
target = ['label']
len(set(predictors) & set(train.columns))
len(predictors)

# train = train[predictors]
train.head()
```

```python
import sys
sys.path.append('/Users/zero/xgboost/python-package')
import xgboost as xgb

from sklearn.model_selection import GridSearchCV
from sklearn.cross_validation import KFold
from sklearn.metrics import log_loss
```

```python
train['label'][train.label == -1] = 0
# train['label'][train.label == 0] = -1

test = train[train.label == -99]
train = train[train.label != -99]

train.shape
test.shape
```

```python
#cv
params = {
    'objective': 'binary:logistic',
    'eta': 0.05,
    'colsample_bytree': 0.886,
    'min_child_weight': 1.6,
    'max_depth': 4,
    'subsample': 0.886,
    'gamma': 0.1,
    'lambda': 10,
    'verbose_eval': True,
    'eval_metric': 'auc',
    'seed': 201803,
    'missing': 0
}

len(predictors)

xgbtrain = xgb.DMatrix(train[predictors], train[target])

cv = xgb.cv(
    params,
    dtrain=xgbtrain,
    num_boost_round=10000,
    nfold=3,
    verbose_eval=100,
    early_stopping_rounds=50,
    seed=201803)
print(cv.shape[0] + ':' + cv.iloc[-1, :]['test-auc'])
# 0.722848(8:2) 0.05*2684
# 0.727476(5:5) 0.05*4193
```

```python
#train/test
params = {
    'objective': 'binary:logistic',
    'eta': 0.05,
    'colsample_bytree': 0.886,
    'min_child_weight': 1.6,
    'max_depth': 4,
    'subsample': 0.886,
    'gamma': 0.1,
    'lambda': 10,
    'scale_pos_weight': int(train[train.label == 0].shape[0]/train[train.label == 1].shape[0]),
    'verbose_eval': 1,
    'eval_metric': 'auc',
    'seed': 201803,
    'missing': -1
}

xgbtrain = xgb.DMatrix(train[predictors], train[target])
xgbtest = xgb.DMatrix(test[predictors])

len(predictors)

model = xgb.train(params, xgbtrain, num_boost_round=4193, verbose_eval=100)
y_pred = model.predict(xgbtest)
y_pred = pd.Series(y_pred, name='test')
```
