```python
#time
def timestamp_datetime(value):
    format = '%Y-%m-%d %H:%M:%S'
    value = time.localtime(value)
    dt = time.strftime(format, value)
    return dt


def convert_data(data):
    data['time'] = data.context_timestamp.apply(timestamp_datetime)
    data['day'] = data.time.apply(lambda x: int(x[8:10]))
    data['hour'] = data.time.apply(lambda x: int(x[11:13]))
    user_query_day = data.groupby(
        ['user_id',
         'day']).size().reset_index().rename(columns={
             0: 'user_query_day'
         })
    data = pd.merge(data, user_query_day, 'left', on=['user_id', 'day'])
    user_query_day_hour = data.groupby(
        ['user_id', 'day',
         'hour']).size().reset_index().rename(columns={
             0: 'user_query_day_hour'
         })
    data = pd.merge(
        data, user_query_day_hour, 'left', on=['user_id', 'day', 'hour'])
    return data


df = convert_data(df)
```

```python
#18、19、20、21、22天做历史数据，23、24天做训练和验证数据
his = df[(df.day != 23) & (df.day != 24) & (df.is_trade != -1)]
train = df[(df.day == 23) | (df.day == 24) | (df.is_trade == -1)]

# del df

his.shape
train.shape
his_trade = his[his.is_trade == 1]

#(357103, 37)
#(121035, 37) 无test
```

```python
#user_shop_count user_item_count user_context_count
user_shop_count = his.groupby(
    ['user_id', 'shop_id'], as_index=False)['user_id'].agg({
        'user_shop_count':
        'count'
    })
train = train.merge(user_shop_count, on=['user_id', 'shop_id'], how='left')

user_item_count = his.groupby(
    ['user_id', 'item_id'], as_index=False)['user_id'].agg({
        'user_item_count':
        'count'
    })
train = train.merge(user_item_count, on=['user_id', 'item_id'], how='left')

user_context_count = his.groupby(
    ['user_id', 'context_id'], as_index=False)['user_id'].agg({
        'user_context_count':
        'count'
    })
train = train.merge(
    user_context_count, on=['user_id', 'context_id'], how='left')

#trade
user_shop_trade_count = his_trade.groupby(
    ['user_id', 'shop_id'], as_index=False)['user_id'].agg({
        'user_shop_trade_count':
        'count'
    })
train = train.merge(
    user_shop_trade_count, on=['user_id', 'shop_id'], how='left')

user_item_trade_count = his_trade.groupby(
    ['user_id', 'item_id'], as_index=False)['user_id'].agg({
        'user_item_trade_count':
        'count'
    })
train = train.merge(
    user_item_trade_count, on=['user_id', 'item_id'], how='left')

user_context_trade_count = his_trade.groupby(
    ['user_id', 'context_id'], as_index=False)['user_id'].agg({
        'user_context_trade_count':
        'count'
    })
train = train.merge(
    user_context_trade_count, on=['user_id', 'context_id'], how='left')

train[[
    'user_shop_count', 'user_item_count', 'user_context_count',
    'user_shop_trade_count', 'user_item_trade_count',
    'user_context_trade_count'
]] = train[[
    'user_shop_count', 'user_item_count', 'user_context_count',
    'user_shop_trade_count', 'user_item_trade_count',
    'user_context_trade_count'
]].fillna(0)
```

```python
#user_brand_count user_city_count
user_brand_count = his.groupby(
    ['user_id', 'item_brand_id'], as_index=False)['user_id'].agg({
        'user_brand_count':
        'count'
    })
train = train.merge(
    user_brand_count, on=['user_id', 'item_brand_id'], how='left')

user_city_count = his.groupby(
    ['user_id', 'item_city_id'], as_index=False)['user_id'].agg({
        'user_city_count':
        'count'
    })
train = train.merge(
    user_city_count, on=['user_id', 'item_city_id'], how='left')

#trade
user_brand_trade_count = his_trade.groupby(
    ['user_id', 'item_brand_id'], as_index=False)['user_id'].agg({
        'user_brand_trade_count':
        'count'
    })
train = train.merge(
    user_brand_trade_count, on=['user_id', 'item_brand_id'], how='left')

user_city_trade_count = his_trade.groupby(
    ['user_id', 'item_city_id'], as_index=False)['user_id'].agg({
        'user_city_trade_count':
        'count'
    })
train = train.merge(
    user_city_trade_count, on=['user_id', 'item_city_id'], how='left')

train[[
    'user_brand_count', 'user_city_count', 'user_brand_trade_count',
    'user_city_trade_count'
]] = train[[
    'user_brand_count', 'user_city_count', 'user_brand_trade_count',
    'user_city_trade_count'
]].fillna(0)
```

```python
#item_occupation_count item_age_count item_gender_count
item_occupation_count = his.groupby(
    ['item_id', 'user_occupation_id'], as_index=False)['item_id'].agg({
        'item_occupation_count':
        'count'
    })
train = train.merge(
    item_occupation_count, on=['item_id', 'user_occupation_id'], how='left')

item_age_count = his.groupby(
    ['item_id', 'user_age_level'], as_index=False)['item_id'].agg({
        'item_age_count':
        'count'
    })
train = train.merge(
    item_age_count, on=['item_id', 'user_age_level'], how='left')

item_gender_count = his.groupby(
    ['item_id', 'user_gender_id'], as_index=False)['item_id'].agg({
        'item_gender_count':
        'count'
    })
train = train.merge(
    item_gender_count, on=['item_id', 'user_gender_id'], how='left')

#trade
item_occupation_trade_count = his_trade.groupby(
    ['item_id', 'user_occupation_id'], as_index=False)['item_id'].agg({
        'item_occupation_trade_count':
        'count'
    })
train = train.merge(
    item_occupation_trade_count,
    on=['item_id', 'user_occupation_id'],
    how='left')

item_age_trade_count = his_trade.groupby(
    ['item_id', 'user_age_level'], as_index=False)['item_id'].agg({
        'item_age_trade_count':
        'count'
    })
train = train.merge(
    item_age_trade_count, on=['item_id', 'user_age_level'], how='left')

item_gender_trade_count = his_trade.groupby(
    ['item_id', 'user_gender_id'], as_index=False)['item_id'].agg({
        'item_gender_trade_count':
        'count'
    })
train = train.merge(
    item_gender_trade_count, on=['item_id', 'user_gender_id'], how='left')

train[[
    'item_occupation_count', 'item_age_count', 'item_gender_count',
    'item_occupation_trade_count', 'item_age_trade_count',
    'item_gender_trade_count'
]] = train[[
    'item_occupation_count', 'item_age_count', 'item_gender_count',
    'item_occupation_trade_count', 'item_age_trade_count',
    'item_gender_trade_count'
]].fillna(0)
```

```python
#shop_score = (star + review_num * review_positive) * service * delivery * description
train.shop_star_level = train.shop_star_level - min(train.shop_star_level)

train['shop_score'] = (
    train.shop_star_level +
    train.shop_review_num_level * train.shop_review_positive_rate
) * train.shop_score_service * train.shop_score_delivery * train.shop_score_description
```

```python
train.context_page_id = train.context_page_id - min(train.context_page_id)
```

```python
#item_score = (sales + collected + pv)
#item_score2 = (sales + collected + pv) / price
#item_score3 = ((sales + collected + pv) / price) / (page + 1)

train['item_score'] = (
    train.item_sales_level + train.item_collected_level * train.item_pv_level)
train['item_score2'] = (
    train.item_sales_level + train.item_collected_level * train.item_pv_level
) / train.item_price_level
train['item_score3'] = (
    (train.item_sales_level + train.item_collected_level * train.item_pv_level
     ) / train.item_price_level) / (train.context_page_id + 1)
```

```python
def pred_split(x):
    pred = []
    tmp = x.split(';')
    for i in range(0, len(tmp)):
        pred.append(tmp[i].split(':'))
    return pred


def predict_category_split(x):
    pred = []
    for i in range(0, len(x)):
        pred.append(x[i][0])
    return pred


def predict_property_split(x):
    pred = []
    for i in range(0, len(x)):
        for j in range(1, len(x[i])):
            if (x[i][j] != '-1'):
                pred.append(x[i][j])
    return pred
```

```python
train[
    'predict_category_property_split'] = train.predict_category_property.apply(
        lambda x: pred_split(x))

train['predict_category_split'] = train.predict_category_property_split.apply(
    lambda x: predict_category_split(x))

train['predict_property_split'] = train.predict_category_property_split.apply(
    lambda x: predict_property_split(x))
```

```python
def item_pred_category_score(x):
    return len(
        list(
            set(x.item_category_split).intersection(
                set(x.predict_category_split))))


def item_pred_property_score(x):
    return len(
        list(
            set(x.item_property_split).intersection(
                set(x.predict_property_split))))
```

```python
train['pred_category_split_count'] = train.predict_category_split.apply(
    lambda x: len(x))
train['pred_property_split_count'] = train.predict_property_split.apply(
    lambda x: len(x))

train['item_pred_category_score'] = train.apply(
    item_pred_category_score, axis=1)
train['item_pred_property_score'] = train.apply(
    item_pred_property_score, axis=1)

train['item_pred_category_score_item%'] = train.item_pred_category_score / (
    train.item_category_split_count + 1) * 10
train['item_pred_property_score_item%'] = train.item_pred_property_score / (
    train.item_property_split_count + 1) * 10

train['item_pred_category_score_pred%'] = train.item_pred_category_score / (
    train.pred_category_split_count + 1) * 10
train['item_pred_property_score_pred%'] = train.item_pred_property_score / (
    train.pred_property_split_count + 1) * 10
```

```python
#item_hot user_hot shop_hot city_hot occupation_hot brand_hot
item_hot = his.groupby(
    ['item_id'], as_index=False)['item_id'].agg({
        'item_hot': 'count'
    })
train = train.merge(item_hot, on=['item_id'], how='left')

user_hot = his.groupby(
    ['user_id'], as_index=False)['user_id'].agg({
        'user_hot': 'count'
    })
train = train.merge(user_hot, on=['user_id'], how='left')

shop_hot = his.groupby(
    ['shop_id'], as_index=False)['shop_id'].agg({
        'shop_hot': 'count'
    })
train = train.merge(shop_hot, on=['shop_id'], how='left')

brand_hot = his.groupby(
    ['item_brand_id'], as_index=False)['item_brand_id'].agg({
        'brand_hot':
        'count'
    })
train = train.merge(brand_hot, on=['item_brand_id'], how='left')

occupation_hot = his.groupby(
    ['user_occupation_id'], as_index=False)['user_occupation_id'].agg({
        'occupation_hot':
        'count'
    })
train = train.merge(occupation_hot, on=['user_occupation_id'], how='left')

city_hot = his.groupby(
    ['item_city_id'], as_index=False)['item_city_id'].agg({
        'city_hot': 'count'
    })
train = train.merge(city_hot, on=['item_city_id'], how='left')

#trade
item_trade_hot = his.groupby(
    ['item_id'], as_index=False)['is_trade'].agg({
        'item_trade_hot': 'sum'
    })
train = train.merge(item_trade_hot, on=['item_id'], how='left')

user_trade_hot = his.groupby(
    ['user_id'], as_index=False)['is_trade'].agg({
        'user_trade_hot': 'sum'
    })
train = train.merge(user_trade_hot, on=['user_id'], how='left')

shop_trade_hot = his.groupby(
    ['shop_id'], as_index=False)['is_trade'].agg({
        'shop_trade_hot': 'sum'
    })
train = train.merge(shop_trade_hot, on=['shop_id'], how='left')

brand_trade_hot = his.groupby(
    ['item_brand_id'], as_index=False)['is_trade'].agg({
        'brand_trade_hot': 'sum'
    })
train = train.merge(brand_trade_hot, on=['item_brand_id'], how='left')

occupation_trade_hot = his.groupby(
    ['user_occupation_id'], as_index=False)['is_trade'].agg({
        'occupation_trade_hot':
        'sum'
    })
train = train.merge(
    occupation_trade_hot, on=['user_occupation_id'], how='left')

city_trade_hot = his.groupby(
    ['item_city_id'], as_index=False)['is_trade'].agg({
        'city_trade_hot': 'sum'
    })
train = train.merge(city_trade_hot, on=['item_city_id'], how='left')

train[[
    'item_hot', 'user_hot', 'shop_hot', 'brand_hot', 'occupation_hot',
    'city_hot', 'item_trade_hot', 'user_trade_hot', 'shop_trade_hot',
    'brand_trade_hot', 'occupation_trade_hot', 'city_trade_hot'
]] = train[[
    'item_hot', 'user_hot', 'shop_hot', 'brand_hot', 'occupation_hot',
    'city_hot', 'item_trade_hot', 'user_trade_hot', 'shop_trade_hot',
    'brand_trade_hot', 'occupation_trade_hot', 'city_trade_hot'
]].fillna(0)
```

```python
#平均消费能力
user_sell_power = his.groupby(
    ['user_id'], as_index=False)['item_price_level'].agg({
        'user_sell_power':
        'sum'
    })
train = train.merge(user_sell_power, on=['user_id'], how='left')

train[['user_sell_power']] = train[['user_sell_power']].fillna(0)

train['user_sell_power_mean'] = train.user_sell_power / (train.user_hot + 1)
train['user_sell_power_mean_item_minus'] = abs(
    train.user_sell_power_mean - train.item_price_level)

#trade
user_sell_trade_power = his_trade.groupby(
    ['user_id'], as_index=False)['item_price_level'].agg({
        'user_sell_trade_power':
        'sum'
    })
train = train.merge(user_sell_trade_power, on=['user_id'], how='left')

train[['user_sell_trade_power']] = train[['user_sell_trade_power']].fillna(0)

train['user_sell_trade_power_mean'] = train.user_sell_trade_power / (
    train.user_trade_hot + 1)
train['user_sell_trade_power_mean_item_minus'] = abs(
    train.user_sell_trade_power_mean - train.item_price_level)
```
