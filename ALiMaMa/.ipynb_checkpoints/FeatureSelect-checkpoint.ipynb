```python
params = {
    'objective': 'binary:logistic',
    'eta': 0.1,
    'colsample_bytree': 0.886,
    'min_child_weight': 1.6,
    'max_depth': 4,
    'subsample': 0.886,
    'gamma': 0.1,
    'lambda': 10,
    'verbose_eval': True,
    'eval_metric': 'auc',
    'seed': 201803,
    'missing': -1
}

len(predictors_zc)
feature_performance = {}

for i in tqdm(list(set(predictors) - set(predictors_zc))):
    temp_features = predictors_zc.copy()
    temp_features.append(i)
    temp_features = list(set(temp_features))
    
    xgbtrain = xgb.DMatrix(train[temp_features], train['is_trade'])
    xgbval = xgb.DMatrix(val[temp_features])
    model = xgb.train(params, xgbtrain, num_boost_round=100)
    y_pred = model.predict(xgbval)
    y_pred = pd.Series(y_pred, name='val')

#     print(str(i) + ':' + str(log_loss(val['is_trade'], y_pred)))
    feature_performance[i] = log_loss(val['is_trade'], y_pred)
```
