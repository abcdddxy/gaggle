```python
train_query = pd.read_csv('./data/new/oppo_round1_train_20180929.txt', names=['prefix', 'query_prediction', 'title', 'tag', 'label'], sep='\t', usecols=['query_prediction'])
val_query = pd.read_csv('./data/new/oppo_round1_vali_20180929.txt', names=['prefix', 'query_prediction', 'title', 'tag', 'label'], sep='\t', usecols=['query_prediction'])
test_query = pd.read_csv('./data/new/oppo_round1_test_A_20180929.txt', names=['prefix', 'query_prediction', 'title', 'tag'], sep='\t', usecols=['query_prediction'])

df_query = pd.concat([train_query, val_query], axis=0)
df_query = pd.concat([df_query, test_query], axis=0)
df_query.reset_index(drop=True, inplace=True)
df_query.columns = ['query_prediction_origin']
df_query.shape
```

```python
df = pd.read_csv('./feature/df_preprocess_andtest.csv', encoding='utf-8')
df =  pd.concat([df, df_query], axis=1)

df.shape
```

```python
df.prefix = df.prefix.fillna('no_pre')
```

```python
%%time
df['is_first_rate'] = df.apply(lambda x: 1 if x.dict_len != 0 and list(eval(x.query_prediction_origin).keys())[0] == x.title else 0, axis=1)
df['title_equal_prefix'] = df.apply(lambda x: 1 if x.prefix == x.title else 0, axis=1)
df['prefix_in_title'] = df.apply(lambda x: 1 if x.prefix in x.title else 0, axis=1)
df['title_startswith_prefix'] = df.apply(lambda x: 1 if x.title.startswith(x.prefix) else 0, axis=1)
df['title_endswith_prefix'] = df.apply(lambda x: 1 if x.title.endswith(x.prefix) else 0, axis=1)
df['rate_big5'] = df.apply(lambda x: 1 if x.title_rate >= 0.5 else 0, axis=1)
df['title_in_query'] = df.apply(lambda x: 1 if x.title in eval(x.query_prediction) else 0, axis=1)

df['title_rank_in_query'] = df.apply(lambda x: list(eval(x.query_prediction).keys()).index(x.title)+1 if x.title_in_query == 1 else 11, axis=1)
df['title_rank_in_query_origin'] = df.apply(lambda x: list(eval(x.query_prediction_origin).keys()).index(x.title)+1 if x.title_in_query == 1 else 11, axis=1)
df['rank_first_origin_rate'] = df.apply(lambda x: float(list(eval(x.query_prediction_origin).values())[0]) if x.dict_len != 0 else 0.6, axis=1)
df['rank_first_origin_rate_sub'] = df.apply(lambda x: x.title_rate - x.rank_first_origin_rate, axis=1)
```

```python
df.shape
df.tail()
```

```python
df.to_csv('./feature/df_feature2.csv', index=None, encoding='utf-8')
```

## ctr

```python
df = pd.read_csv('./feature/df_feature2.csv', encoding='utf-8')

df.shape
```

```python
items = ['prefix', 'title', 'tag']

for item in tqdm_notebook(items):
    temp = df[df.is_val == 0].groupby(item, as_index = False)['label'].agg({item + '_click':'sum', item + '_ctr':'mean', item + '_count':'count'})
    df = pd.merge(df, temp, on=item, how='left')
    
for i in tqdm_notebook(range(len(items))):
    for j in range(i + 1, len(items)):
        item_g = [items[i], items[j]]
        temp = df[df.is_val == 0].groupby(item_g, as_index=False)['label'].agg({'_'.join(item_g) + '_click': 'sum', '_'.join(item_g) + '_ctr': 'mean', '_'.join(item_g) + '_count':'count'})
        df = pd.merge(df, temp, on=item_g, how='left')
```

```python
df.shape
```

```python
df.to_csv('./feature/df_feature3_ctr.csv', index=None, encoding='utf-8')
```

## 符号数字

```python
import urllib
import re
```

```python
df = pd.read_csv('./feature/df_feature4_ctr_extra.csv', encoding='utf-8', usecols=['is_val', 'prefix', 'title', 'query_prediction'])

df.shape
```

```python
%%time
df['query_values_sum'] = df.query_prediction.apply(lambda x: sum(np.array(list(eval(x).values())).astype(float)))
```

```python
# df['title_url'] = df.title.apply(lambda x: urllib.parse.unquote(x))
# df['prefix_has_url'] = df.apply(lambda x: 0 if x.title_url == x.title else 1, axis=1)

df['prefix_has_num'] = df.prefix.apply(lambda x: 0 if re.search('\d', x) == None else 1)
df['prefix_has_letter'] = df.prefix.apply(lambda x: 0 if re.search('\w', x) == None else 1)
df['title_has_num'] = df.title.apply(lambda x: 0 if re.search('\d', x) == None else 1)
df['title_has_letter'] = df.title.apply(lambda x: 0 if re.search('\w', x) == None else 1)
df['title_url_has_num'] = df.title_url.apply(lambda x: 0 if re.search('\d', x) == None else 1)
df['title_url_has_letter'] = df.title_url.apply(lambda x: 0 if re.search('\w', x) == None else 1)
```

```python
df['prefix_inter_title_len'] = df.apply(lambda x: len(set(x.prefix).intersection(set(x.title))), axis=1)
df['prefix_inter_title_url_len'] = df.apply(lambda x: len(set(x.prefix).intersection(set(x.title_url))), axis=1)

df['prefix_inter_title_len_rate_prefix'] = df.apply(lambda x: x.prefix_inter_title_len / (len(set(x.prefix))+1), axis=1)
df['prefix_inter_title_len_rate_title'] = df.apply(lambda x: x.prefix_inter_title_len / (len(set(x.title))+1), axis=1)
df['prefix_inter_title_url_len_rate_prefix'] = df.apply(lambda x: x.prefix_inter_title_url_len / (len(set(x.prefix))+1), axis=1)
df['prefix_inter_title_url_len_rate_title_url'] = df.apply(lambda x: x.prefix_inter_title_url_len / (len(set(x.title_url))+1), axis=1)
```

```python
df = df.drop(['is_val', 'prefix', 'title', 'title_url', 'query_prediction'], axis=1)
```

```python
df.to_csv('./feature/feature_rank_query_re.csv', index=None)
```
