```python
df = pd.read_csv('./data/df_preprocess.csv')

df.shape
df.head()
```

```python
df = df.sort_values(by=['time'], ascending=True)
```

```python
def timestamp_datetime(value):
    format = '%Y-%m-%d %H:%M:%S'
    value = time.localtime(value)
    dt = time.strftime(format, value)
    return dt


def convert_data(data):
    data['time'] = data.time.apply(timestamp_datetime)
    data['day'] = data.time.apply(lambda x: int(x[8:10]))
    data['hour'] = data.time.apply(lambda x: int(x[11:13]))
    data['minute'] = data.time.apply(lambda x: int(x[14:15])) * 10
    return data


# train 6.1 6.2 test 6.3
df = convert_data(df)
```

## 预处理

```python
from collections import Counter
```

```python
# 去除出现次数少的
make_most = [
    i[0]
    for i in sorted(
        Counter(df.make).items(), key=lambda x: x[1], reverse=True)[:210]
]
model_most = [
    i[0]
    for i in sorted(
        Counter(df.model).items(), key=lambda x: x[1], reverse=True)[:1150]
]
adid_most = [
    i[0]
    for i in sorted(
        Counter(df.adid).items(), key=lambda x: x[1], reverse=True)[:650]
]
orderid_most = [
    i[0]
    for i in sorted(
        Counter(df.orderid).items(), key=lambda x: x[1], reverse=True)[:530]
]
creative_id_most = [
    i[0]
    for i in sorted(
        Counter(df.creative_id).items(), key=lambda x: x[1], reverse=True)[:
                                                                           350]
]
app_id_most = [
    i[0]
    for i in sorted(
        Counter(df.app_id).items(), key=lambda x: x[1], reverse=True)[:150]
]
inner_slot_id_most = [
    i[0]
    for i in sorted(
        Counter(df.inner_slot_id).items(), key=lambda x: x[1], reverse=True)
    [:260]
]
```

```python
df.make = df.make.apply(lambda x: x if x in make_most else 9999)
df.model = df.model.apply(lambda x: x if x in model_most else 9999)
df.adid = df.adid.apply(lambda x: x if x in adid_most else 9999)
df.orderid = df.orderid.apply(lambda x: x if x in orderid_most else 9999)
df.creative_id = df.creative_id.apply(lambda x: x if x in creative_id_most else 9999)
df.app_id = df.app_id.apply(lambda x: x if x in app_id_most else 9999)
df.inner_slot_id = df.inner_slot_id.apply(lambda x: x if x in inner_slot_id_most else 9999)
```

```python
full = df.copy()
his = df[df.time < '2039-06-01 00:00:00'].copy()
df = df[df.time >= '2039-06-01 00:00:00'].copy()

his.shape
df.shape

full['noclick'] = 1 - full.click
his['noclick'] = 1 - his.click
df['noclick'] = 1 - df.click
```

```python
train = df[df.click != -1]
test = df[df.click == -1]
```

```python
ad_col = [
    'adid', 'advert_id', 'orderid', 'campaign_id', 'creative_id',
    'creative_tp_dnf', 'creative_type', 'advert_name',
    'advert_industry_inner_split_0', 'advert_industry_inner_split_1'
]
ad_cate_col = [
    'creative_is_jump', 'creative_is_download', 'creative_is_js',
    'creative_is_voicead', 'creative_has_deeplink'
]
medim_col = ['app_cate_id', 'app_id', 'inner_slot_id', 'app_paid']
user_col = [
    'user_tags_0', 'user_tags_1', 'user_tags_2', 'user_tags_3', 'user_tags_4',
    'user_tags_5', 'user_tags_6', 'user_tags_7', 'user_tags_8', 'user_tags_9',
    'user_tags_10', 'user_tags_11', 'user_tags_12', 'user_tags_13',
    'user_tags_14', 'user_tags_15', 'user_tags_16', 'user_tags_17',
    'user_tags_18', 'user_tags_19', 'user_tags_20', 'user_tags_21',
    'user_tags_22', 'user_tags_23', 'user_tags_24', 'user_tags_25',
    'user_tags_26', 'user_tags_27', 'user_tags_28', 'user_tags_29',
    'user_tags_30', 'user_tags_31', 'user_tags_32', 'user_tags_33',
    'user_tags_34', 'user_tags_35', 'user_tags_36', 'user_tags_37'
]
context_col = [
    'city', 'province', 'carrier', 'devtype', 'make', 'model', 'nnt', 'os',
    'osv', 'os_name'
]

cont_col = ['creative_width', 'creative_height']
by_col = ['instance_id', 'time']
target = ['click']

cross_col = ad_col + ad_cate_col + medim_col + user_col + context_col
full_col = ad_col + ad_cate_col + medim_col + user_col + context_col + cont_col + by_col + target
len(full_col)
```

## 单特征

```python
train.shape
test.shape
```

```python
for i in tqdm_notebook(cont_col + ad_cate_col + user_col):
    for j in ['mean', 'var', 'count']:
        tmp = his.groupby([i], as_index=False)['click'].agg({'click_' + i + '_' + j: j})
        train = train.merge(tmp, on=['click'], how='left')
        tmp = full.groupby([i], as_index=False)['click'].agg({'click_' + i + '_' + j: j})
        test = test.merge(tmp, on=['click'], how='left')
    train['click_' + i + '_mean_sub'] = abs(train[i] - train['click_' + i + '_mean'])
    test['click_' + i + '_mean_sub'] = abs(test[i] - test['click_' + i + '_mean'])

for i in tqdm_notebook(ad_col + ad_cate_col + medim_col + user_col + context_col):
    tmp = his.groupby([i], as_index=False)['click'].agg({i + '_pos_sample': 'sum'})
    train = train.merge(tmp, on=[i], how='left')
    tmp = his.groupby([i], as_index=False)['noclick'].agg({i + '_nag_sample': 'sum'})
    train = train.merge(tmp, on=[i], how='left')
    train[i + 'pos/nag_rate'] = (train[i + '_pos_sample'] + 1) / (train[i + '_nag_sample'] + 1)
    
    tmp = full.groupby([i], as_index=False)['click'].agg({i + '_pos_sample': 'sum'})
    test = test.merge(tmp, on=[i], how='left')
    tmp = full.groupby([i], as_index=False)['noclick'].agg({i + '_nag_sample': 'sum'})
    test = test.merge(tmp, on=[i], how='left')
    test[i + 'pos/nag_rate'] = (test[i + '_pos_sample'] + 1) / (test[i + '_nag_sample'] + 1)
```

## 交叉特征

```python
from itertools import product, permutations
```

```python
len(list(permutations(cross_col, 2)))
```

```python
train = train[full_col]
test = test[full_col]

train.shape
test.shape
```

```python
order = 0

for (i, j) in tqdm_notebook(list(permutations(cross_col, 2))):
    tmp_pos = his.groupby([i, j], as_index=False)['click'].agg({i + '_' + j + '_pos_sample': 'sum'})
    tmp_nag = his.groupby([i, j], as_index=False)['noclick'].agg({i + '_' + j + '_nag_sample':'sum'})
    tmp = tmp_pos.merge(tmp_nag, on=[i, j], how='left')
    tmp[i + '_' + j + 'pos/nag_rate'] = (tmp[i + '_' + j + '_pos_sample'] + 1) / (tmp[i + '_' + j + '_nag_sample'] + 1)
    tmp = tmp[[i, j, i + '_' + j + 'pos/nag_rate']]
    train = train.merge(tmp, on=[i, j], how='left')
    
    tmp_pos = full.groupby([i, j], as_index=False)['click'].agg({i + '_' + j + '_pos_sample': 'sum'})
    tmp_nag = full.groupby([i, j], as_index=False)['noclick'].agg({i + '_' + j + '_nag_sample':'sum'})
    tmp = tmp_pos.merge(tmp_nag, on=[i, j], how='left')
    tmp[i + '_' + j + 'pos/nag_rate'] = (tmp[i + '_' + j + '_pos_sample'] + 1) / (tmp[i + '_' + j + '_nag_sample'] + 1)
    tmp = tmp[[i, j, i + '_' + j + 'pos/nag_rate']]
    test = test.merge(tmp, on=[i, j], how='left')
    if train.shape[1] >= 500:
        train.to_csv('./feature/feature_nonoise/cross_feature_train' + str(order) + '.csv', index=None)
        test.to_csv('./feature/feature_nonoise/cross_feature_test' + str(order) + '.csv', index=None)
        print('finish file ' + str(order) + ' ' + str(train.shape) + ' ' + str(test.shape))
        order += 1
        train = train[full_col]
        test = test[full_col]
    del tmp
    del tmp_pos
    del tmp_nag
    _ = gc.collect()

train.to_csv('./feature/feature_nonoise/cross_feature_train' + str(order) + '.csv', index=None)
test.to_csv('./feature/feature_nonoise/cross_feature_test' + str(order) + '.csv', index=None)
print('finish file ' + str(order) + ' ' + str(train.shape) + ' ' + str(test.shape))
```

## 时序特征

```python
del his
_ = gc.collect()
```

```python
# 某一时间段内点击次数/点击转化率
for time in [['day'], ['day', 'hour'], ['day', 'hour', 'minute']]:
    for i in tqdm_notebook(ad_col + medim_col + context_col):
        tmp_by = time + [i]
        tmp_full = df.groupby(
            tmp_by, as_index=False)['click'].agg({
                i + '_' + '_'.join(time) + '_query':
                'count'
            })
        df = df.merge(tmp_full, on=tmp_by, how='left')

        tmp_pos = df.groupby(
            tmp_by, as_index=False)['click'].agg({
                i + '_' + '_'.join(time) + '_posquery':
                'sum'
            })
        tmp = tmp_full.merge(tmp_pos, on=tmp_by, how='left')
        tmp[i + '_' + '_'.join(time) + '_posrate'] = (
            tmp[i + '_' + '_'.join(time) + '_posquery'] + 1) / (
                tmp[i + '_' + '_'.join(time) + '_query'] + 1)
        
        tmp = tmp[tmp_by + [i + '_' + '_'.join(time) + '_posrate']]
        df = df.merge(tmp, on=tmp_by, how='left')
        
        del tmp_full
        del tmp_pos
        del tmp
        _ = gc.collect()
```

## USER_ID

```python
import hashlib
```

```python
train = pd.read_csv('./data/round1_iflyad_train.txt', sep='\t')
test = pd.read_csv('./data/round1_iflyad_test_feature.txt', sep='\t')

train.shape
train.head()
test.shape
test.head()
```

```python
test['click'] = -1
df = pd.concat([train, test], axis=0)
df.reset_index(drop=True, inplace=True)

df.shape
```

```python
df = df.sort_values(by=['time'], ascending=True)
```

```python
userid_col = [
    'city', 'province', 'carrier', 'devtype', 'make', 'model', 'nnt', 'os',
    'osv', 'os_name', 'user_tags'
]

def user_id_str(x):
    id_str = ''
    for i in userid_col:
        id_str += '+' + str(x[i])
    return id_str

df['user_id_str'] = df.apply(user_id_str, axis=1)
```

```python
def user_id_md5(x):
    if type(x) == float:
        x = ' '
    hl = hashlib.md5()
    hl.update(x.encode(encoding='utf-8'))
    return hl.hexdigest()

df['user_id'] = df.user_id_str.apply(user_id_md5)
df['user_id2'] = df.user_tags.apply(user_id_md5)
```
